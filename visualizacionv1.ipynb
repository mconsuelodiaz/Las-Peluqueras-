```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
file_path = '/content/Basededatos_munoz.csv'
data = pd.read_csv(file_path)

# Mostrar las primeras filas del DataFrame
print(data.head())

# Graficar los datos en un gráfico de barras
ax = data.plot(kind='bar', figsize=(12, 8), colormap='viridis')  # Ajustar el tamaño y el color

# Personalizar el gráfico
plt.title('Grafico de Barras de los Datos', fontsize=16)
plt.xlabel('Índice', fontsize=14)
plt.ylabel('Valores', fontsize=14)

# Añadir una leyenda con título
plt.legend(title='Columnas', bbox_to_anchor=(1.05, 1), loc='upper left')

# Añadir etiquetas a cada barra
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ["Más cómodo", "Más rápido", "Más económico", "No tener otra opción", "Otro"]
percentages = [59.7, 29.0, 12.7, 4.4, 3.6]

# Create pie chart
plt.pie(percentages, labels=labels, autopct="%1.1f%%")

# Add title
plt.title("¿Por qué factores son más importantes para ti al elegir un servicio de delivery?")

# Show the plot
plt.show()
```
